```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import cross_validation
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn import svm

ds = pd.read_csv("G:\\UIT BS-SE Semester-VI Course Folders\\CS 321 - Artificial Intelligence\Lab\\A.I. Lab 10\\data_banknote_authentication.txt")
x = ds.iloc[:,0:4].values
y = ds.iloc[:,4].values

X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.3,random_state=109)

model = svm.SVC(kernel='linear')
model.fit(x,y)
predicted = model.predict(X_test)
print("Predicted Value for 30:70 split:", predicted)
print("Accuracy Score:",accuracy_score(Y_test, predicted))
print("Confusion Matrix:\n",confusion_matrix(Y_test, predicted))
print(classification_report(Y_test, predicted))

model = cross_validation.cross_val_score(model,x,y,cv=5)
print(model)

X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.5,random_state=42)
model = svm.SVC(kernel='linear')
model.fit(x,y)
predicted = model.predict(X_test)
print("Predicted Value for 50:50 split:", predicted)
print("\nAccuracy Score:",accuracy_score(Y_test, predicted))
print("\nConfusion Matrix:\n",confusion_matrix(Y_test, predicted))
print(classification_report(Y_test, predicted))

model = cross_validation.cross_val_score(model,x,y,cv=5)
print(model)

X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.7,random_state=17)
model = svm.SVC(kernel='linear')
model.fit(x,y)
predicted = model.predict(X_test)
print("Predicted Value for 70:30 split:", predicted)
print("Accuracy Score:",accuracy_score(Y_test, predicted))
print("\nConfusion Matrix:\n",confusion_matrix(Y_test, predicted))
print(classification_report(Y_test, predicted))

model = cross_validation.cross_val_score(model,x,y,cv=5)
print(model)

X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.9,random_state=133)
model = svm.SVC(kernel='linear')
model.fit(x,y)
predicted = model.predict(X_test)
print("Predicted Value for 90:10 split:", predicted)
print("Accuracy Score:",accuracy_score(Y_test, predicted))
print("\nConfusion Matrix:\n",confusion_matrix(Y_test, predicted))
print(classification_report(Y_test, predicted))

model = cross_validation.cross_val_score(model,x,y,cv=5)
print(model)





```
