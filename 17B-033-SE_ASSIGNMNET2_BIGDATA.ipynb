```python
import sys
introfile=open("17B-033-SE_ASSIGNMNET2_BIGDATA.txt","r")
file_text = introfile.read()
print("TEXT IN FILE")
print(file_text )
print("Total character in a file is:")
print(len(file_text) )
print("WELCOM TO SHINGLING \nCHARACTER BASED TOKENS(PRESS 1)\nWORDS BASED TOKENS(PRESS 2)\nBYTE SIZE BASED TOKENS(PRESS 3)\n")
print("Enter your choice :")
choice=int(input())
if  choice==1:
    print("Shingling by character\n")
    print("Enter shingling length :")
    shingleLength = int(input())
    print("\nTxt after shingling is:")
    for i in range(len(file_text)  - shingleLength + 1):
        print (file_text [i:i+shingleLength])
                   
elif choice==2: 
    print("Shingling by word\n")
    print("Enter shingling length :")
    shingleLength = int(input())
    split_text = file_text.split()
    print("\nThe split of sentence is:")
    print(split_text)
    print("\nSets after shingling is:")
    for i in range(len(split_text) - shingleLength + 1):
        if len(split_text[i]) < len(file_text):
            print (split_text[i:i+shingleLength])   
            
elif choice==3: 
    print("Shingling by byte size\n")       
    print("Byte size is:")
    print(sys.getsizeof(file_text))
    print("Enter byte size length :")
    shingleLength = int(input())
    split_text = file_text.split()
    print("ASCII code of text is:") 
    txt=[]        
    for i in split_text:
        txt.extend(ord(num) for num in i) 
    print(str(txt))
    print("\nThe split of sentence is:")
    print(split_text)
    print("\nBytes after shingling is:")
    for i in range(len(txt) - shingleLength + 1): 
        if len(txt) < len(file_text): 
            print(txt[i:i+shingleLength])
            
else:
    print("INVALID INPUT")
```
